````python
# Sin conprehension
def initialize_inventory(products):
    inventory = {}
    for product in products:
        quantity = int(input(f"Enter the quantity for {product}: "))
        inventory[product] = quantity
    return inventory

def get_customer_orders():
    customer_orders = set()
    while True:
        product = input("Enter the product name (or 'done' to finish): ")
        if product.lower() == 'done':
            break
        customer_orders.add(product)
    return customer_orders

def update_inventory(customer_orders, inventory):
    for order in customer_orders:
        if order in inventory:
            inventory[order] -= 1
    return inventory

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    unique_products_ordered = len(customer_orders.intersection(products))
    percentage_unique = (unique_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_unique

def print_order_statistics(order_statistics):
    total_products_ordered, percentage_unique = order_statistics
    print(f"Total products ordered: {total_products_ordered}")
    print(f"Percentage of unique products ordered: {percentage_unique:.2f}%")

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# Main program execution
products = ["apple", "banana", "orange", "grape"]
inventory = initialize_inventory(products)
customer_orders = get_customer_orders()
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
````

````python
def initialize_inventory(products):
    return {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}

def get_customer_orders(num_orders):
    return {input(f"Enter the name of a product that a customer wants to order: ") for _ in range(num_orders)}

def update_inventory(customer_orders, inventory):
    return {product: (quantity - 1 if product in customer_orders else quantity) for product, quantity in inventory.items()}

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique = (total_products_ordered / len(products)) * 100 if products else 0
    return total_products_ordered, percentage_unique

def calculate_total_price(customer_orders, prices):
    return sum(prices[product] for product in customer_orders)

def print_order_statistics(order_statistics):
    [print(f"{label}: {value}") for label, value in zip(
        ["Total Products Ordered", "Percentage of Unique Products Ordered"],
        [order_statistics[0], f"{order_statistics[1]:.2f}%"]
    )]

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]

def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    inventory = initialize_inventory(products)
    
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = get_customer_orders(num_orders)
    
    inventory = update_inventory(customer_orders, inventory)
    
    order_statistics = calculate_order_statistics(customer_orders, products)
    print("\nOrder Statistics:")
    print_order_statistics(order_statistics)
    
    print("\nUpdated Inventory:")
    print_updated_inventory(inventory)
    
    prices = {product: float(input(f"Enter the price of {product}: ")) for product in customer_orders}
    total_price = calculate_total_price(customer_orders, prices)
    print(f"Total Price: {total_price:.2f}")

# Ejecutar el programa principal
main()
````
